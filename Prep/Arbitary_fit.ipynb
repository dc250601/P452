```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
```

```python
x = 3*np.random.random(1000)
y = np.sin(x**2)
```

```python
plt.scatter(x,y)
```

```python
def f0(x):
    return x**0
def f1(x):
    return x
def f2(x):
    return (3*(x**2) - 1)/2
def f3(x):
    return 0.5*(5*(x**3) - 3*x)
def f4(x):
    return (1/8)*(35*(x**4) - 30*(x**2) +3)
F = [f0,f1,f2,f3,f4]
```

```python
alpha = np.zeros(5)

for i in range(10):
    F_x = np.array([(alpha[i]*F[i](x)) for i in range(len(F))]).sum(0)
    r = np.sum(y - F_x)
    psi = [F[i](x).sum() for i in range(len(F))]
    alpha = alpha - r/psi
```
